```python
import pandas as pd
from sqlalchemy import create_engine
```

### Store Beers CSV into DataFrame

```python
csv_file = "Resources/beers.csv"
beers_df = pd.read_csv(csv_file)
beers_df.head()
```

### Clean Data

```python
#delete unecessary column
new_beers_df = beers_df.drop(["notes"], axis=1) 
#select US beer only
new_beers_us_df=new_beers_df[new_beers_df["country"]=="US"]
#reset index
new_beers_us_df=new_beers_us_df.reset_index(drop=True)
new_beers_us_df.head()
```

```python
#check if there is any repeat id if not use it as primary key of beer 
# new_beers_us_df["id"].value_counts()
```

### Store Breweries CSV into DataFrame

```python
csv_file = "Resources/breweries.csv"
breweries_df = pd.read_csv(csv_file)
breweries_df.head()
```

### Clean data

```python
new_breweries_df =breweries_df.drop(["notes"],axis=1)
# change id name 
new_breweries_df = new_breweries_df.rename(columns={"id":"brewery_id"})
#select US beer only
new_breweries_us_df=new_breweries_df[new_breweries_df["country"]=="US"]
new_breweries_us_df.head()
#check repeat items 
# new_breweries_us_df["brewery id"].value_counts()
```

### Store Review Data into CSV

```python
csv_file = "Resources/beer_reviews.csv"
review_df = pd.read_csv(csv_file)
review_df.head()
```

```python
review_df.info()
```

```python
#drop row if beer id is not in beer data frame
beear_id_list=new_beers_us_df["id"].tolist()
review_clean_df=review_df[~review_df.beer_beerid.isin(beear_id_list)]
#check if there is any row dropped
review_clean_df.info()
```

### Create new data with select columns

### Beer ID Table

```python
beer_id_table=new_beers_us_df[["id","name"]].copy()
beer_id_table.head()
```

### Brewery ID Table

```python
brewery_id_table = new_breweries_us_df[["brewery_id","name"]].copy()
brewery_id_table.head()
```

### Beer Profile Table

```python
beer_profile_table = new_beers_us_df[["id","name","style","abv","availability"]].copy()
beer_profile_table.head()
```

### Brewery Location info

```python
brewery_location_df=new_breweries_us_df[["brewery_id","name","city","state","country"]].copy()
brewery_location_df.head()
```

### Beer Reviews

```python
review_table=review_clean_df[["beer_beerid","beer_name","review_overall","review_aroma","review_appearance","review_taste"]]
review_table.head()
```

### Connect to local database

```python
from sqlalchemy import create_engine
```

```python
rds_connection_string = "postgres:200233032p@localhost:5432/ETL_Project"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

### Use pandas to load csv converted DataFrame into database

```python
beer_id_table.to_sql(name='beer_id_table', con=engine, if_exists='append', index=False)
brewery_id_table.to_sql(name='brewery_id_table', con=engine, if_exists='append', index=False)
beer_profile_table.to_sql(name='beer_profile_table', con=engine, if_exists='append', index=False)
```
