```python
import pandas as pd
from sqlalchemy import create_engine
# import datetime
from tabulate import tabulate
import tabulatehelper as th
```

### Store Beers CSV into DataFrame

```python
csv_file = "Resources/beers.csv"
beers_df = pd.read_csv(csv_file)
#check df content data type and non-null object
print(beers_df.info())
#display dataframe
beers_df.head()
```

### Clean Data

```python
#drop unecessary column
new_beers_df = beers_df.drop(["notes"], axis=1) 
#select US beer only
new_beers_us_df = new_beers_df[new_beers_df["country"]=="US"]
#check beer name repeat times
print(new_beers_us_df["name"].value_counts().head(10))
# #check beer name repeat times
# print(new_beers_us_df["id"].value_counts().head(10))
#drop duplicate ID if there is a repeat
new_beers_us_df=new_beers_us_df.drop_duplicates(subset='id', keep='first')
#rename columns for better understanding
new_beers_us_df = new_beers_us_df.rename(columns={"abv":"alchol_by_volume","style":"beer_style" })
#reset index
new_beers_us_df = new_beers_us_df.reset_index(drop=True)
new_beers_us_df.head()
```

### Store Breweries CSV into DataFrame

```python
csv_file = "Resources/breweries.csv"
breweries_df = pd.read_csv(csv_file)
print(breweries_df.info())
breweries_df.head()
```

### Clean data

```python
new_breweries_df =breweries_df.drop(["notes"],axis=1)
# change column name 
new_breweries_df = new_breweries_df.rename(columns={"id":"brewery_id","name":"brewery_name"})
#drop duplicate ID if there is a repeat
new_breweries_df.drop_duplicates(subset='brewery_id', keep='first')
#select US beer only
new_breweries_us_df=new_breweries_df[new_breweries_df["country"]=="US"]
new_breweries_us_df.head()
```

### Store Review Data into CSV

```python
csv_file = "Resources/beer_reviews.csv"
review_df = pd.read_csv(csv_file)
print(review_df.info())
review_df.head()
```

```python
# change column name 
review_df = review_df.rename(columns={"id":"brewery_id","beer_name":"name","beer_abv":"alchol_by_volume","beer_beerid":"id"})
# change review _time to panda date format
review_df['review_time'] = pd.to_datetime(review_df['review_time'],unit='s')
review_df.head()
```

```python
#Since review data does not specify beer production country, so we only select beer id in selected US beer from beer dataframe
beer_id_list=new_beers_us_df["id"].tolist()
new_review_df=review_df[review_df.id.isin(beer_id_list)]
#check if there is any row dropped
new_review_df.head()
```

### Beer Reviews

```python
review_table = new_review_df[["id","name","beer_style","alchol_by_volume","review_overall","review_aroma","review_appearance","review_taste","review_time"]].copy()
review_table.head()
```

### Adjusting Table

```python
# Creating binning for very low, low, average, high, and very high alcohol concentrations
bins = [0,2,4,6,8,50]
labels = ['Very Low', 'Low', 'Average', 'High', 'Very High']
adjust_review_df = new_review_df.copy()
adjust_review_df['alcohol_by_volume_category'] = pd.cut(adjust_review_df['alchol_by_volume'], bins=bins, labels=labels)
```

```python
# Creating adjusted review_overall scores by accounting for potential review bias
# Calculate review_overall_profilename_avg, join to existing dataframe, save to new dataframe
df_adj = adjust_review_df.join(adjust_review_df.groupby('review_profilename')['review_overall'].mean(), on='review_profilename', rsuffix='_profilename_avg')
# Calculate bias
df_adj['bias'] = df_adj['review_overall'].mean() - df_adj['review_overall_profilename_avg']
# Calculate adjusted scores
df_adj['review_overall_adjusted'] = df_adj['review_overall'] + df_adj['bias']
# Same for review_aroma
df_adj = df_adj.join(df_adj.groupby('review_profilename')['review_aroma'].mean(), on='review_profilename', rsuffix='_profilename_avg')
df_adj['bias_aroma'] = df_adj['review_aroma'].mean() - df_adj['review_aroma_profilename_avg']
df_adj['review_aroma_adjusted'] = df_adj['review_aroma'] + df_adj['bias_aroma']
# Same for review_appearance
df_adj = df_adj.join(df_adj.groupby('review_profilename')['review_appearance'].mean(), on='review_profilename', rsuffix='_profilename_avg')
df_adj['bias_appearance'] = df_adj['review_appearance'].mean() - df_adj['review_appearance_profilename_avg']
df_adj['review_appearance_adjusted'] = df_adj['review_appearance'] + df_adj['bias_appearance']
# Same for review_palate
df_adj = df_adj.join(df_adj.groupby('review_profilename')['review_palate'].mean(), on='review_profilename', rsuffix='_profilename_avg')
df_adj['bias_palate'] = df_adj['review_palate'].mean() - df_adj['review_appearance_profilename_avg']
df_adj['review_palate_adjusted'] = df_adj['review_palate'] + df_adj['bias_palate']
# Same for review_taste
df_adj = df_adj.join(df_adj.groupby('review_profilename')['review_taste'].mean(), on='review_profilename', rsuffix='_profilename_avg')
df_adj['bias_taste'] = df_adj['review_taste'].mean() - df_adj['review_taste_profilename_avg']
df_adj['review_taste_adjusted'] = df_adj['review_taste'] + df_adj['bias_taste']
df_adj.columns
```

```python
#Select the adjusted columns from df_adj
adjusted_table = df_adj[['id', 'review_overall_adjusted', 'review_aroma_adjusted', 'review_appearance_adjusted', 'review_palate_adjusted', 'review_taste_adjusted']].copy()
adjusted_table.head()
```
